```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import random
import warnings

warnings.filterwarnings(action='ignore')
```

```python
train_data = pd.read_csv('/content/removed_train1.csv')
test_data = pd.read_csv('/content/removed_test1.csv')
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
X_train = train_data.drop(columns = ['Diagnosis'])
X_test = test_data.drop(columns = ['Diagnosis'])

y_train = train_data['Diagnosis']
y_test = test_data['Diagnosis']
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold, cross_val_score
```

```python
train_data.head()
```

```python
lgb_clf = lgb.LGBMClassifier(n_estimators = 100, learning_rate = 0.1, max_depth = 7, random_state = 42)
lgb_clf.fit(X_train, y_train)

predictions = lgb_clf.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f'정확도 : {accuracy}')
```

```python
k_fold = KFold(n_splits = 5, shuffle = True, random_state = 42)
scores = cross_val_score(lgb_clf, X_train, y_train, cv = k_fold)

for i, score in enumerate(scores, 1):
    print(f"Fold {i} Score: {score}")

print(f"Mean Score: {scores.mean()}")
```
